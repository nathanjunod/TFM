**Domain Architecture of ADGRE2 (Q9UHX3)**

This notebook visualizes the domain structure of the human ADGRE2 protein, a member of the adhesion GPCR family. The protein plays a role in immune responses and is implicated in inflammatory disorders such as vibratory urticaria.

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Domain data
domains = [
    ("EGF-like 1", 25, 66),
    ("EGF-like 2 (Ca-binding)", 67, 118),
    ("EGF-like 3 (Ca-binding)", 119, 162),
    ("EGF-like 4 (Ca-binding)", 163, 211),
    ("EGF-like 5 (Ca-binding)", 212, 260),
    ("GAIN domain", 354, 530),
    ("GPS region", 482, 530),
]

protein_length = 912

# Choose a colormap
cmap = cm.get_cmap("tab10", len(domains))  # or "Set3", "tab20", etc.

# Create plot
fig, ax = plt.subplots(figsize=(16, 4))
ax.hlines(1, 0, protein_length, color="gray", linewidth=5)

# Plot domains with distinct colors
for i, (name, start, end) in enumerate(domains):
    color = cmap(i)
    ax.hlines(1, start, end, color=color, linewidth=12)
    ax.text((start + end) / 2, 1.08, name, ha='center', fontsize=12, rotation=45, color=color)

# N- and C-terminus
ax.text(0, 1.18, "N", ha="center", va="center", fontsize=14, fontweight="bold")
ax.text(protein_length, 1.18, "C", ha="center", va="center", fontsize=14, fontweight="bold")

# Tick marks
for pos in range(0, protein_length + 1, 100):
    ax.text(pos, 0.93, str(pos), ha='center', fontsize=10)

# Layout
ax.set_xlim(0, protein_length)
ax.set_ylim(0.9, 1.25)
ax.axis('off')
ax.set_title("Domain Architecture of ADGRE2 (Q9UHX3)", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```
